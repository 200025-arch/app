```python
df = pd.read_csv('data.csv', sep=",")

#Affiche les 5 premiers enregistrements (lignes)
```

```python
#handling missing values

#J'obtiens "true" où il y'a des valeurs manquantes
df.isnull()
```

```python
df.isnull().any()
```

```python
#J'obtiens le nombre de valeurs manquantes par colonnes
df.isnull().sum()
```

```python
#fill the missing values

df_filled = df.fillna(0)
df_filled
```

```python
## Filling missing values with the mean of the colonne
df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
#Renaming columns
#ça fonctionne comme un dictionnaire ('clé' : 'valeur')
df = df.rename(columns={'Date' : 'Sales Date', 'Value' : 'Sales Value'})
df.head()
```

```python
#Change datatypes

df['Value_new'] = df['Sales Value'].fillna(df['Sales Value'].mean()).astype(int)
df.head()
```

```python
#Appliquer une fonction
#On utilise "apply" pour appliquer une fonction

df['Double Value'] = df['Sales Value'].apply(lambda x:x*2)
df.head()
```

```python
def times_3(x):
    return x*3

df['Triple Value'] = df['Sales Value'].apply(times_3)
df.head()
```

```python
## Data aggregating and grouping
df.head()
```

```python
#La moyenne des ventes par goupe de produit
#Ici j'ai groupé par produit et j'ai récupéré la moyenne des ventes

grouped_mean = df.groupby('Product')['Sales Value'].mean()
grouped_mean
```

```python
#Le total des ventes par produit et pas région
grouped_sum = df.groupby(['Product', 'Region'])['Sales Value'].sum()
grouped_sum
```

```python
#aggregate multiple functions

#On obtient la moyenne, la somme et ne total de transactions par région
grouped_agg = df.groupby('Region')['Sales Value'].agg(['mean', 'sum', 'count'])
grouped_agg
```

```python
#Merging and joining dataframes
df1 = pd.DataFrame({'Key' : ['A', 'B', 'C'], 'Value' : [1,2,3]})
df2 = pd.DataFrame({'Key' : ['A', 'B', 'D'], 'Value' : [4,5,6]})

print(df1)
print(df2)
```

```python
df1
```

```python
df2
```

```python
#Merge Dataframe on the 'Key columns'

pd.merge(df1, df2, on='Key', how="inner")
```

```python
pd.merge(df1, df2, on='Key', how="outer")
```

```python
pd.merge(df1, df2, on='Key', how="left")
```

```python
pd.merge(df1, df2, on='Key', how="right")
```
